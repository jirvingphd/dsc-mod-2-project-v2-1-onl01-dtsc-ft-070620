```python
def plot_columns(df, col = 'sqft_above',target= 'price'):
    
    fig, axes = plt.subplots(figsize=(10,6),ncols=2)
    sns.boxplot(x=col,data=df,ax=axes[0])
    sns.regplot(x=col,y=target, data=df, ax=axes[1])
    plt.tight_layout()
    return fig, axes
```

```python
def find_outiers_Z(col):
    zscores = np.abs(stats.zscore(col))
    idx_outliers = zscores > 3
    return idx_outliers
```

```python
idx_outs_br = find_outiers_Z(df['bedrooms'])
idx_outs_br
```

```python
plot_columns(df[~idx_outs_br],'bedrooms')
```
